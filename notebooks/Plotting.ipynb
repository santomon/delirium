##PCA vs no PCA plot

```
import seaborn as sns
import matplotlib.pyplot as plt
pca_plotter = pc.Plotter()
pca_plotter.load_corrs("torch_inference", "deeplabv3_resnet50", True, True, False, [3, 4], "/content/gdrive/MyDrive/results3")
pca_plotter.load_corrs("torch_inference", "deeplabv3_resnet50", False, True, False, [3, 4], "/content/gdrive/MyDrive/results3")

# custom mod for pca plotting
def plot_average_for_all_subjects(data, hue_order, plot, hue="model_name", **kwargs):

    if plot == sns.barplot or plot == "bar":
        plot = sns.barplot
    elif plot == sns.violinplot or plot == 'violin':
        plot = sns.violinplot
    elif plot == sns.boxplot or plot =="box":
        plot = sns.boxplot
    else:
        print("plot should be one of bar, violin or box")

    if 'palette' not in kwargs.keys():
        kwargs['palette'] = sns.color_palette("colorblind")

    fig = plt.figure(figsize=((30, 15)))
    ax = fig.add_subplot(1, 1, 1)
    plot(
        x="ROI",
        y="correlation",
        hue=hue,
        hue_order=hue_order,
        data=data,
        ax=ax,
        **kwargs
    )

    plt.xlabel("ROI", fontsize=30)
    plt.ylabel("Correlations (r)", fontsize=30)
    handles, _ = ax.get_legend_handles_labels()
    ax.legend(title="DeeplabV3 ResNet50", handles=handles, labels=["with PCA", "without PCA"])
    # fig.savefig("effects_of_pca_vs_nopca.pgf")
    fig.savefig("effects_of_pca_vs_nopca.png")

    plt.show()
plot_average_for_all_subjects(pca_plotter.data, [True, False], "box", "did_pca")
```
